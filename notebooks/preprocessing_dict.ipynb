baseline — переведем по первому словарю, который нашли в интернете. Почистим его, чтобы можно было пользоваться.

```python
import re
import numpy as np
import pickle
```

```python
with open("..\data\dict.txt", mode="r", encoding="UTF-8") as f:
    text = f.readlines()
```

```python
text[:20]
```

```python
len(text)
```

Почистим

```python
text = [x.replace("\n","") for x in text if (x != "\n") & (x != "\t") & (re.match("^[а-яА-я\s0-9]*$", x) == None)]
```

```python
len(text)
```

```python
text
```

```python
ru_dict_raw = [x.split('\t') for x in text]
```

```python
add = [x for x in ru_dict_raw if len(x) == 1]
ru_dict_raw = [x for x in ru_dict_raw if len(x) > 1]
```

```python
add = [x[0].split(' ') for x in add]
add = [x for x in add if len(x) > 1]
```

```python
ru_dict_raw = ru_dict_raw + add
```

```python
ru_dict_raw
```

```python
ru_zh = {}
zh_ru = {}

for line in ru_dict_raw:
    ru_words = []
    zh_words = []
    for words in line:
        is_ru = bool(re.match("^[а-яА-Я\s.,:;?!ё()-–]*$", words))
        
        if is_ru:
            ru_words += [x.strip() for x in re.split("[,;–]", words) if (len(x.strip()) > 0) & (x.strip() != "-a") & (x.strip() != "-ы")]
        else:
            zh = [x.strip() for x in re.split("[,;–，]", words)]
            if zh:
                zh_words += zh
                
    for r in ru_words:
        ru_zh[r] = zh_words
        
    for z in zh_words:
        zh_ru[z] = ru_words
        
```

```python
zh_ru
```

```python
pickle.dump(ru_zh, open("ru_zh.pkl", "wb"))
pickle.dump(zh_ru, open("zh_ru.pkl", "wb"))
```
